# Grid
Funkce grid nám umožní vykreslit mřížku na vykreslovaný plot

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Data ze sportovních hodinek")
plt.xlabel("Průměrný tep")
plt.ylabel("Spálené kalorie")

plt.plot(x, y)

plt.grid()

plt.show()
```

Můžeme specifikovat jaké osy chceme vykreslit

```python
plt.plot(x, y)

plt.grid(axis = 'x')

plt.show()
```

# Subplot
Funkce subplot nám umožňuje vykreslit více grafů na jeden plot

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

Funkce subplot bere 3 argumenty:
- rozměr na ose x
- rozměr na ose y
- index, který plot vykreslujeme

```python
plt.subplot(1, 2, 1)
# tento subplot obsahuje 1 řádek, 2 sloupce a vykreslujeme plot 1

plt.subplot(1, 2, 2)
# tento subplot obsahuje 1 řádek, 2 sloupce a vykreslujeme plot 2
```

pokud otočíme předchozí příklad

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)
plt.plot(x,y)

plt.show()
```

Subplot může teoreticky obsahovat nekonečné množství plotů. Problém nastane s přehledností, protože s každým dalším plotem se zmenší velikost zbylých plotů.

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 1)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 2)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 3)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 4)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 5)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 6)
plt.plot(x,y)

plt.show()
```

# Scatter
Scatter je funkce která vykreslí bod na grafu ke každému záznamu. Tato fuknce bere 2 pole o stejné délce, jedno pro osu x a jedno pro osu y

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
```

Příklad nahoře může být výsledek pozorování 13 projíždějících aut.

Osa x je stáří auta a osa y rychlost jako projeli

```python
plt.scatter(x, y)
plt.xlabel("Stáří auta [roky]")
plt.ylabel("Rychlost [km/s]")
plt.show()
```

Výsledek z tohoto pozorování může být, že čím novější auto tím rychleji jelo.

To ale může být náhoda, máme poměrně malý vzorek dat (13 aut)

```python
#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)

#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y)

plt.xlabel("Stáří auta [roky]")
plt.ylabel("Rychlost [km/s]")

plt.show()
```

S dalšími daty z nového dne, tentokrát máme 15 vzorků, můžeme poměrně s jistotou říct, že nová auta jezdí rychleji.

## ColorMap
Colormapa je barevný rozsah, krerý nabíváhodnot od 0 do 100.

Colormapu nastavíme argumentem cmap

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

# musíme vytvořit pole v rozsahu 0-100, počet prvků = len(x)
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')

# plt.colorbar()
plt.show()
```

## Alfa kanál
Pomocí alfa kanálu jsme schopni přidat do grafů prvek transparentnosti. To se udělpomocí argumentu alpha, tento argument může nabívat hodnot od 0 do 1

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes, alpha=0.5)

plt.show()
```

# Ukázka z praxe
Zadání:
V souboru (Prumerna_Teplota.csv), naleznete 3 sloupce s hodnotami průměrné denní teploty pro každý den v roce.
Úkolem je zpracování dat pomocí scriptu v Pythonu pomocí knihovny MatPlotLib a vytvoření grafů tak, že:
- Obrázek bude vhodně rozdělen tak, aby byly zobrazeny jednotlivé grafy pro každý měsíc, kde bude vykreslena závislost průměrné denní teploty na jednotlivých dnech daného měsíce, budou vykreslena měsíční průměrná teplota

```python
data = np.genfromtxt("Prumerna_Teplota.csv", delimiter=";", dtype=str)
headers = data[0]
data = data[1::]

pocet_dni_v_mesici = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
mesice = ["Leden", "Únor", "Březen", "Duben", "Květen", "Červen", "Červenec", "Srpen", "Září", "Říjen", "Listopad", "Prosinec"]

#def get_teploty(pocet_dni:int, )

# hlídáme si správnou pozici odkud máme číst z "dat"
iter_count = 0
# pro každý měsíc
for index_mesice, pocet_dni in enumerate(pocet_dni_v_mesici):
    # získáme data v rozsahu počtu dnů
    zaznam_dat = data[iter_count:pocet_dni+iter_count]
    
    teploty = []
    # pro každý ze získaných záznamů
    for index_dnu, den in enumerate(zaznam_dat):
        # vybereme pouze teplotu
        teploty.append(float(zaznam_dat[index_dnu].split(",")[2]))
    # vytvoříme si pole 1 až počet dnů
    dny = [i for i in range(1, pocet_dni+1)]
    # vypočítáme průměr teplot
    prumer_teplot = np.average(teploty)

    # vytvoříme subplot 3x4
    plt.subplot(3, 4, index_mesice+1)
    # vykreslíme naměřené teploty
    plt.plot(np.array(dny), np.array(teploty))
    # vykreslíme průměrnou teplotu
    plt.plot(np.array(dny), np.full(pocet_dni, prumer_teplot))
    # přidáme popisky
    plt.title(mesice[index_mesice])
    plt.xlabel("Den")
    plt.ylabel("Teplota [°C]")

    iter_count += pocet_dni

plt.subplots_adjust(left=0.2, bottom=0.2, right=2, top=2, wspace=0.4, hspace=0.4)
plt.show()
```
